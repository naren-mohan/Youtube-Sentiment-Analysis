```python
nexttoken = ""
videoID = "6i-cagHpBcs"
```

```python
comment = []
author = []
like_count = []
reply_count = []

nextPage = True

nexttoken = ""
videoID = "6i-cagHpBcs"

while(nextPage):
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=videoID,
        order="relevance",
        maxResults=5682,
        pageToken=nexttoken
    )
    response = request.execute()
    
    if 'nextPageToken' in response.keys():
        nexttoken = response['nextPageToken']
    else:
        nextPage = False
    
    for item in response['items']:
        comment.append(item['snippet']['topLevelComment']['snippet']['textOriginal'])
        author.append(item['snippet']['topLevelComment']['snippet']['authorDisplayName'])
        like_count.append(item['snippet']['topLevelComment']['snippet']['likeCount'])
        reply_count.append(item['snippet']['totalReplyCount'])
    
print(len(comment))
```

```python
data = {'Comments': comment,
       'Author': author,
       'LikeCount': like_count,
       'ReplyCount': reply_count}
df = pd.DataFrame(data=data)
```
